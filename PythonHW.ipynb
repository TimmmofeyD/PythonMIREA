## 1 Работа

```python
from math import acos, sqrt


def main(z) -> float: 
    return (
    sqrt(91 * z ** 3 - z ** 2 - z)
    + 36 * acos(z) ** 6
    - (z ** 7) / ((z + 1 + 63 * z ** 3) ** 5 - 22 * (z ** 2 + 87 + z ** 3))
)
```

```python
main(0.84)
```

##  3 Задание

```python
from math import tan


def main(n, a, m, b) -> float:
    return (
        sum(1 - (47 * i ** 2 - 89 * i ** 3 - i) / 63 - i ** 7
            for i in range(1, n + 1))
        -
        sum(
            12 * tan(c) ** 2 + i ** 7 + 25 * (2 + k) ** 3
            for k in range(1, b + 1)
            for c in range(1, m + 1)
            for i in range(1, a + 1)
        )
    )
```

```python
main(3,3,3,2)
```
